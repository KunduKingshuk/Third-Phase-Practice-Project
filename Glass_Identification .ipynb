## Look at variable relation to Glass type

RI and Na to glass type

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='1',y='1.52101',hue='1.1',data=df)
```

####  Na is an important feature for glass type


##### Ri distribution does not provide much insight.

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='13.64',y='4.49',hue='1.1',data=df)
```

#### Higher Glass type have lower levels of 13.64

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='71.78',y='0.06',hue='1.1',data=df)
```

##### 71.78 is of moderate importance in deciding glass type

#### All types of glass have lower values of 0.06

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='8.75',y='0.00',hue='1.1',data=df)
```

#### Higher glass type usually have higher values of 0.00

##### 8.75 helps moderately in identifying the glass type

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='4.49',y='8.75',hue='1.1',data=df)
```

##### 4.49 plays a vital role in identifying glass type.

## Checking Distribution of features of importance

```python
plt.figure(figsize=(15,15))

plt.subplot(2,2,1)
sns.violinplot(x='1.1',y='4.49',data=df)

plt.subplot(2,2,2)
sns.violinplot(x='1.1',y='0.00',data=df)

plt.subplot(2,2,3)
sns.violinplot(x='1.1',y='13.64',data=df)

plt.subplot(2,2,4)
sns.violinplot(x='1.1',y='1.52101',data=df)
```

#### Not much insight into distribution if we look at violin plots

#####   Confirming observation using correlation

```python
corr = df.corr() 
corr
```

#### Confirmed Observations:

#### 1.52101 (Na), 4.49 (Al), 13.64 (Mg) and 0.00 (Ba) does helps in identifying glass types.
#### 71.78 (K), 8.75(Ca) does not provide much insight into glass types.
